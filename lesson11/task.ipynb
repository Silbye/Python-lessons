```python
x = sympy.Symbol('x')
print(type(x))
```

```python
f_x = -12 * x **4 * sympy.sin(sympy.cos(x)) - 18 * x ** 3 + 5 * x ** 2 + 10 * x - 30
print(f_x)
```

```python
f_diff_x = f_x.diff()
```

```python
plot(f_x)
```

```python
plot(f_diff_x, xlim=(-11, 11))
```
